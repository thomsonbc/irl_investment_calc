```python
def pivotIndex(nums: list[int]) -> int:
    idx=-1
    for i in range(len(nums)):
        if i == 0:
                if 0 == sum(nums[1:-1]):
                    idx=i
                    break
        if sum(nums[0:i]) == sum(nums[i:-1]):
            idx=i
            break
    return idx
```

```python
pivotIndex(nums=nums)
```
